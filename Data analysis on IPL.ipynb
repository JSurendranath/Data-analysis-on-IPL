```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('matches.csv')
df1 = pd.read_csv('deliveries.csv')
```

```python
# Count missing values for each column
missing_values = df.isnull().sum()

# Print the result
print(missing_values)
```

```python
# Group by city and count the number of seasons for each city
seasons_per_city = df.groupby('city').size()

# Find the city with the maximum number of seasons
most_seasons_city = seasons_per_city.idxmax()

print("City with the most number of seasons:", most_seasons_city)
```

```python
# Group by year and count the number of matches for each year
matches_per_year = df.groupby('season').size()

# Find the year with the maximum number of matches
most_matches_year = matches_per_year.idxmax()

print("Year with the most number of matches:", most_matches_year)
```

```python
# Filter the data for Mumbai Indians in 2011
mi_2011 = df[(df['team1'] == 'Mumbai Indians') | (df['team2'] == 'Mumbai Indians') & (df['season'] == 2011)]

# Count the number of wins
max_wins_mi_2011 = mi_2011['winner'].value_counts().idxmax()

print("Maximum wins by Mumbai Indians in 2011:", max_wins_mi_2011)
```

```python
# Plotting a pie chart for 'toss_decision'
toss_decision_counts = df['toss_decision'].value_counts()
labels = toss_decision_counts.index
sizes = toss_decision_counts.values

plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Toss Decision')
plt.show()
```

```python
# Count the number of wins based on toss decisions
toss_decision_wins = df.groupby(['toss_winner', 'toss_decision'])['winner'].count()

print(toss_decision_wins)
```

```python
# Filter data for matches where teams batting second won
batting_second_wins = df[df['win_by_wickets'] > 0]

# Group by year and count the number of matches won batting second
batting_second_wins_by_year = batting_second_wins.groupby('season').size()

# Print the result
print(batting_second_wins_by_year)
```

```python
# column name 'batsman' for player names and 'total_runs' for the total runs scored
top_10_batsmen = df1.groupby('batsman')['total_runs'].sum().nlargest(10)

# Plotting a line plot for the top 10 run-getters
top_10_batsmen.plot(kind='line', marker='o', figsize=(10, 6))
plt.title('Consistency Among Top 10 Run-Getters')
plt.xlabel('batsman')
plt.ylabel('total_runs')
plt.show()
```

```python
# Filter data for the year 2019 and non-zero boundaries (6's and 4's)
filtered_data = df1[(df['season'] == 2019) & ((df1['batsman_runs'] == 4) | (df1['batsman_runs'] == 6))]

# Group by batsman and sum the runs
runs_by_batsman = filtered_data.groupby('batsman')['batsman_runs'].sum()

# Find the batsman with the highest runs
highest_runs_batsman = runs_by_batsman.idxmax()

print("Batsman with the highest number of runs by hitting 6's and 4's in 2019:", highest_runs_batsman)
```

```python
# Group by bowler and count the number of wickets
wickets_by_bowler = df1.groupby('bowler')['player_dismissed'].count()

# Find the bowler with the most wickets
most_wickets_bowler = wickets_by_bowler.idxmax()

print("Bowler with the most number of wickets:", most_wickets_bowler)
```

```python
# Filter data for Virat Kohli in 2016
kohli_2016 = df1[(df1['batsman'] == 'Virat Kohli') & (df['season'] == 2016)]

# Calculate total runs and total balls faced
total_runs = kohli_2016['batsman_runs'].sum()
total_balls_faced = kohli_2016['ball'].count()

# Calculate the strike rate
strike_rate = (total_runs / total_balls_faced) * 100

print("Virat Kohli's strike rate in 2016:", strike_rate)
```

```python
# Group by bowler and sum the extras
extras_by_bowler = df1.groupby('bowler')['extra_runs'].sum()

# Find the bowler with the most extras
most_extras_bowler = extras_by_bowler.idxmax()

print("Bowler with the maximum number of extras:", most_extras_bowler)
```

```python
# Group by venue and count the number of matches
matches_by_venue = df['venue'].value_counts()

# Find the venue with the most matches
most_matches_venue = matches_by_venue.idxmax()

print("Venue that has hosted the most number of IPL matches:", most_matches_venue)
```

```python
# Filter data for the match between Sunrisers Hyderabad and Royal Challengers Bangalore in 2017
match_data = df[(df1['batting_team'] == 'Sunrisers Hyderabad') & (df1['bowling_team'] == 'Royal Challengers Bangalore') & (df['season'] == 2017)]

# Extract the Player of the Match for the specified match
player_of_match = match_data['player_of_match'].iloc[0]

print("Player of the Match in the 2017 match between Sunrisers Hyderabad and Royal Challengers Bangalore:", player_of_match)
```

```python
# Filter data for run-outs
run_outs_data = df1[df1['dismissal_kind'] == 'run out']

# Group by batsman and count the number of run-outs
run_outs_by_batsman = run_outs_data.groupby('batsman')['dismissal_kind'].count()

# Get the top three batsmen with the most run-outs
top_three_run_outs = run_outs_by_batsman.nlargest(3)

print("Top three batsmen with the most number of run-outs across seasons:")
print(top_three_run_outs)
```

```python
# Filter data for matches where V Kohli faced JJ Bumrah
kohli_vs_bumrah = df1[(df1['batsman'] == 'V Kohli') & (df1['bowler'] == 'JJ Bumrah')]

# Calculate the total runs
total_runs_vs_bumrah = kohli_vs_bumrah['batsman_runs'].sum()

print("Total runs scored by V Kohli when the bowler was JJ Bumrah:", total_runs_vs_bumrah)
```

```python
# Filter data for dismissals via caught and bowled
caught_and_bowled_data = df1[df1['dismissal_kind'] == 'caught and bowled']

# Group by batsman and count the number of caught and bowled dismissals
caught_and_bowled_by_batsman = caught_and_bowled_data.groupby('batsman')['dismissal_kind'].count()

# Find the player with the most caught and bowled dismissals
most_caught_and_bowled_player = caught_and_bowled_by_batsman.idxmax()

print("Player with the maximum number of dismissals via caught and bowled across all seasons:", most_caught_and_bowled_player)
```

```python
# Calculate strike rate for each player
player_strike_rate = df1.groupby('batsman')['batsman_runs'].sum() / df1.groupby('batsman')['ball'].count() * 100

# Find the player with the highest strike rate
highest_strike_rate_player = player_strike_rate.idxmax()

print("Player with the highest hard-hitting ability (highest strike rate):", highest_strike_rate_player)
```
